```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

def backward_elimination(X, y, significance_level=0.05):
    num_features = X.shape[1]
    for i in range(num_features):
        regressor = sm.OLS(y, X).fit()
        max_pvalue = max(regressor.pvalues)
        if max_pvalue > significance_level:
            max_pvalue_index = np.argmax(regressor.pvalues)
            X = np.delete(X, max_pvalue_index, 1)
        else:
            break
    return X

# Cargar el conjunto de datos en un DataFrame
df = pd.read_csv('datos.csv')

# Separar las características (X) y la variable objetivo (y)
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# Aplicar la eliminación hacia atrás
X = np.append(arr=np.ones((X.shape[0], 1)).astype(int), values=X, axis=1)
X_opt = backward_elimination(X, y)

# Imprimir las características seleccionadas
selected_features = df.columns[np.nonzero(X_opt[0])[0]-1]
print(selected_features)
```
