```python
# listing nulls
pd.set_option('display.max_rows', None)

nulos_por_columna = df.isnull().sum().sort_values(ascending=False)

print(nulos_por_columna)
```

```python
# null values > 1000
columnas_con_nulos = []
for columna, cantidad in nulos_por_columna.items():
    if cantidad > 1000:
        columnas_con_nulos.append(columna)

print(columnas_con_nulos)
```

```python
# df without too many nulls (5623, 124)
df_sin_nulos = df.drop(columnas_con_nulos, axis=1)
```

```python
# df without any nulls (2805, 124)
df_sin_nulos = df_sin_nulos.dropna()
```
